```python
import pandas as pd
import numpy as np

df = pd.read_csv('datastore/รวมม.csv')
#print(df)
```

```python
df
```

```python
df["tags"] = df["tags"].str.replace("\[|\]|\'", "").str.split(",")
df
```

```python
df2 = pd.DataFrame(columns=['quote','author','tags'])

for index, row in df.iterrows():
    for value in str(row['tags']).split(" "):
        d = {'quote':row['quote'], 'value':value}
        df2 = df2.append(d, ignore_index=True)
df2 = df2.groupby('quote')['value'].value_counts()
df2 = df2.unstack(level=-1).fillna(0)
df2
```

```python
df2.to_csv('datastore/tagsรวม.csv', index = False)
```
